Function and dictionary to fill empty positions of players

```python
def fill_empty_positions(aggregated_data):
    """
    Fill empty positions in the aggregated data with positions from non-empty rows of the same player.

    Parameters:
        aggregated_data (DataFrame): Aggregated data with player-season statistics.

    Returns:
        DataFrame: Aggregated data with filled positions.
    """
    # Group by 'name' and check for non-empty positions
    player_non_empty_positions = aggregated_data.groupby('name')['position'].first().reset_index()

    # Merge back to fill empty positions
    filled_positions = pd.merge(aggregated_data, player_non_empty_positions, on='name', suffixes=('', '_filled'))

    # Replace missing positions with the filled positions
    filled_positions['position'] = filled_positions['position_filled'].fillna(filled_positions['position'])

    # Drop the temporary 'position_filled' column
    filled_positions.drop(columns=['position_filled'], inplace=True)

    return filled_positions
```
